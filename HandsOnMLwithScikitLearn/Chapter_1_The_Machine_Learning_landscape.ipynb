```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

```python
oecd_bli = pd.read_csv('Chapter_1_oecd_bli_2017.csv', thousands=',')
gdp_per_capita = pd.read_table('Chapter_1_WEO_Data.xlsx',thousands=',',delimiter=',',
encoding='latin1', na_values="n/a")
print(oecd_bli['Unit Code'][-5:])
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
print(oecd_bli.head(5))
# oecd_bli = oecd_bli.pivot_table(index="Country", columns="Indicator", values="Value")
# countries = oecd_bli.index.values
# oecd_bli['countries'] = countries

# # gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
# # gdp_per_capita.set_index("Country", inplace=True)
# # gdp_per_capita
```

```python
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot_table(index="Country", columns=["Indicator"], values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    remove_indices = [0, 1, 6, 8, 33, 34, 35]
    keep_indices = list(set(range(36)) - set(remove_indices))
    return full_country_stats[["GDP per capita", 'Life satisfaction','Air pollution','Dwellings without basic facilities','Educational attainment','Employees working very long hours','Employment rate','Feeling safe walking alone at night','Homicide rate','Household net adjusted disposable income','Household net financial wealth','Housing expenditure','Years in education']].iloc[keep_indices]
```

```python
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
```

```python
X = np.c_[country_stats["GDP per capita"]]
y = np.c_[country_stats["Life satisfaction"]]
print(X,y)
```

```python
country = np.array(country_stats.index.values)
lifeSatisfaction = np.array(country_stats[['Life satisfaction']].values)
airPollution = np.array(country_stats[['Air pollution']].values)
dwellings = np.array(country_stats[['Dwellings without basic facilities']].values)
education = np.array(country_stats[['Educational attainment']].values)
work = np.array(country_stats[['Employees working very long hours']].values)
employment = np.array(country_stats[['Employment rate']].values)
safety = np.array(country_stats[['Feeling safe walking alone at night']].values)
homicide = np.array(country_stats[['Homicide rate']].values)
houseHoldNetDisposableIncome = np.array(country_stats[['Household net adjusted disposable income']].values)
houseHoldNetWealth = np.array(country_stats[['Household net financial wealth']].values)
housingExpenditure = np.array(country_stats[['Housing expenditure']].values)
yearsInEducation = np.array(country_stats[['Years in education']].values)
```

```python
df = pd.DataFrame({'Country': country.flatten(), 'LifeSatisfaction': lifeSatisfaction.flatten(),'AirPollution':airPollution.flatten(),'DwellingsWithoutBasicFacilities':dwellings.flatten(),'EducationAttainment':education.flatten(),'WorkForLongHours':work.flatten(),'EmploymentRate':employment.flatten(),'WorkerSafetyAtNight':safety.flatten(),'HomicideRate':homicide.flatten(),'houseHoldNetDisposableIncome':houseHoldNetDisposableIncome.flatten(),'houseHoldNetWealth':houseHoldNetWealth.flatten(),'housingExpenditure':housingExpenditure.flatten(),'yearsInEducation':yearsInEducation.flatten()})
```
